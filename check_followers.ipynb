```python
def get_followers(x):
    print(x,'получаем последователей')
    
    try: 
        user = api.get_user(x)        
        return user.followers_count
    except:
        return None
```

```python
def get_following(x):
    print(x,'получаем последуемых')
    user = api.get_user(x)
    if user.friends_count >= 30:
        return user.friends_count
    else:
        return None
```

```python
def get_tweets(x):
    print(x,'получаем Твиты')
    
    if api.statuses_count(x) >= 10:
        user = api.get_user(x)
        return user.statuses_count
    else:
        return None
```

```python
def get_likes(x):
    print(x,'получаем Лайки')
    user = api.get_user(x)
    
    if user.favourites_count >= 30:
        
        return user.favourites_count
    else:
        return None
```

```python
def get_date(x):
    print(x,'получаем дату')
    user = api.get_user(x)
    if datetime(2017, 1, 1) > user.created_at:
        return 1
    else:
        return None
```

```python
def get_img(x):
    print(x,'получаем рисунок')
    try:
        user = api.get_user(x)
    #     print(user.profile_image_url_https)
    except:
        print(x, 'Не найден')
        return None
    if 'default' in str(user.profile_image_url_https):
         return 1
    else:
        return None
#     print(user.profile_image_url_https)
```
